```python
import pandas as pd
import numpy as np
```

```python
def get_data(year="2015-16"):
    
    ret = pd.read_csv("data/CTSchoolDaySAT" + year + ".csv",skiprows=3,header=[0,1])
    
    return ret

get_data().head()
```

```python
scores_17 = get_data(year="2016-17")
scores_17.head()
```

```python
def make_json(data,subject="ELA",base_data={}):

    ret = data[data["Unnamed: 2_level_0","Subject"] == subject]
    
    json_ret = base_data
    
    
    def num_null(n):
        
        ret = pd.to_numeric(n, errors="coerce")
                
        #print ret, np.isnan(np.array([ret], dtype=np.float64))[0]
        
        
        #print ret, pd.isnull(n), np.isnan(n)
        
        
        if np.isnan(np.array([ret], dtype=np.float64))[0]:
            return None
        
        return ret
    
    def convert_row (row):
                
        school_name = row["Unnamed: 0_level_0","District"]
        
        test_count = row["Unnamed: 4_level_0","Total NumberTested"]
            
        json_ret[school_name + " (" + str(test_count) + ")"] = {
            "name": "Level 1 - Not met",
            "index": 0,
            "data": [num_null(row["Unnamed: 8_level_0","%"])]
           },{
            "name": "Level 2 - Approaching",
            "index": 1,
            "data": [num_null(row["Unnamed: 10_level_0","%"])]
           },{
            "name": "Level 3 - Met",
            "index": 2,
            "data": [num_null(row["Unnamed: 12_level_0","%"])]
           },{
            "name": "Level 4 - Exceeded",
            "index": 3,
            "data": [num_null(row["Unnamed: 14_level_0","%"])]
           },{
            "name": "Met or Exceeded" ,
            "index": 4,
            "data": [num_null(row["Unnamed: 16_level_0","%"])]
        }
                
    for index, row in ret.iterrows():
        convert_row(row)
        
    return json_ret
    
make_json(get_data(year="2016-17"),subject="Math")
```
